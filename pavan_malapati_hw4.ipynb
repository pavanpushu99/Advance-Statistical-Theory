```python
from scipy.stats import norm

mu = 100
sigma = 15

# (i) P(92 < X < 108)
p_i = norm.cdf(108, mu, sigma) - norm.cdf(92, mu, sigma)

# (ii) P(X > 105)
p_ii = 1 - norm.cdf(105, mu, sigma)

# (iii) P(2X + 5 < 200) => P(X < 97.5)
p_iii = norm.cdf((200 - 5) / 2, mu, sigma)

print(f'(i) P(92 < X < 108): {p_i:.4f}')
print(f'(ii) P(X > 105): {p_ii:.4f}')
print(f'(iii) P(2X + 5 < 200): {p_iii:.4f}')
```
