## Exercise 1.3

Generate a sequence of 50 random binary numbers (0, 1), when the likelihood of 1 is p using the command binom.rvs(1, p, size=50).

```python
from scipy.stats import binom
import numpy as np

np.random.seed(1)

# Generate and print 50 random samples from a binomial distribution with n=1 and p=0.5
b_s = binom.rvs(1, 0.5, size=50)
print(b_s)
```

Do this for the values p = 0.1, 0.3, 0.7, 0.9. Count the number of 1s in these random sequences, by summing up the result sequence.

```python
np.random.seed(1)
for p in (0.1, 0.3, 0.7, 0.9):
  o = binom.rvs(1, p, size=50)
  print(p, sum(o))
```

The expected values of the sums are 5, 15, 35 and 45.

The outcomes are 4,12,33 and 43 for p = 0.1,0.3,0.7,0.9.Therefore, The difference between my outcome and the expected values are 1,3,2 and 2.

The following data is the time (in sec.) to get from 0 to 60 mph for a sample of 15 German- and 20 Japanese-made cars.

```python
germanCars = [10, 10.9, 4.8, 6.4, 7.9, 8.9, 8.5, 6.9, 7.1,
5.5, 6.4, 8.7, 5.1, 6.0, 7.5]
japaneseCars = [9.4, 9.5, 7.1, 8.0, 8.9, 7.7, 10.5, 6.5, 6.7,
9.3, 5.7, 12.5, 7.2, 9.1, 8.3, 8.2, 8.5, 6.8, 9.5, 9.7]
```

Compare and contrast the acceleration times of German- and Japanese-made cars, in terms of their five summary statistics.

```python
germanCars = pd.Series(germanCars)
japaneseCars = pd.Series(japaneseCars)

comp_stat = pd.DataFrame({
  'German': germanCars.describe(),
  'Japanese': japaneseCars.describe(),
})
print(comp_stat)
```

German-made cars have a quicker average acceleration time (7.37 seconds) compared to Japanese-made cars (8.46 seconds). The range of times for German cars is also wider, with times going from 4.8 to 10.9 seconds, while Japanese cars range from 5.7 to 12.5 seconds. German cars tend to accelerate faster and have a slightly higher variation in acceleration times. Both types show that most cars fall between the 25th and 75th percentiles, but German cars generally perform better in terms of quicker acceleration.
